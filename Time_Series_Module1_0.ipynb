<h2><u>What is time series analysis?</h2></u>

A time series analysis is a method of studying a collection of data points over a period of time. Instead of capturing data points intermittently or arbitrarily, time series analyzers record data points at constant intervals over a predetermined length of time.

![ts1.2.png](https://cdn.corporatefinanceinstitute.com/assets/Time-Series-Analysis-1024x729.png)

As you can interpret from the above image, in time series the response variable is studied against time. The time can be in days,weeks,months,years or even minutes. Time series analysis can be used to demonstrate how variables change over time. Time is an important variable since it reveals how the data changes over time, as well as the final outcomes. It also guarantees that any trends or patterns found aren't outliers and that seasonal variation is taken into account. Time series data can also be used for forecasting, or anticipating future data based on previous data.

<h2><u>Components of time series</h2></u>

The components of a time series are the many reasons or forces that affect the values of an observation in a time series.

![ts1.5.png](https://www.googleapis.com/download/storage/v1/b/kaggle-user-content/o/inbox%2F3526173%2F9534e2d240cbe31285ba6787b121ea95%2FIntroduction-to-Time-Series-Data.jpg?generation=1595620883447392&alt=media)



Various components of time series are:


1.   Trend:<br>
A time series trend is its long-term pattern. A trend exists when there is a long-term increase or decrease in the data. It does not have to be linear. Sometimes we will refer to a trend as changing direction, when it might go from an increasing trend to a decreasing trend. Depending on whether the time series exhibits an increasing or declining long term pattern, a trend might be positive or negative.
A time series is stationary in the mean if it does not display an increasing or decreasing tendency.

![ts1.6.png](https://productimize.com/wp-content/uploads/2019/08/Component-in-Time-Series_Trend.png)

2.   Seasonality<br>
When seasonal elements such as the time of year or the day of the week affect a time series, it is called a seasonal pattern. Seasonality has a predictable and defined frequency.

![ts1.7.png](https://www.vosesoftware.com/riskwiki/images/image1858.gif)

3.   Cyclicity<br>
When data shows increases and falls that are not of a constant frequency, it is called a cycle. Economic conditions are usually to blame for these variations, which are often linked to the "business cycle." Many individuals mix up cyclic and seasonal behaviour, yet the two are quite different. The variations are cyclic if they do not have a defined frequency; the pattern is seasonal if the frequency is constant and linked to some feature of the calendar.

![ts1.8.png](https://i.ytimg.com/vi/ca0rDWo7IpI/maxresdefault.jpg)

4.  Irregularity<br>
The irregular component arises from non-systematic and, in certain cases, unpredictable short-term variations in a series, such as unusual weather patterns.

![ts1.9.png](https://www.datascienceprophet.com/wp-content/uploads/2020/08/irregular.jpg)

Now its time to actually visualize the data. For that define a plot_df function having attributes like df(dataframe to be plotted), x(columns on X-axis), y(columns on Y-axis), title(title of the graph), xlabel(name given to the X-axis), ylabel(name given to the Y-axis), dpi(dots per inch). <br>
<br><b>Here plot the graph by keeping dates(index) on X-axis and values on Y-axis.</b>

Refer the documentation to know more about matplotlib:https://matplotlib.org/stable/tutorials/index

```python
# Draw Plot
# define plot_df function for plotting
# parameters: df, x, y, title="", xlabel='Date', ylabel='Value', dpi=100

def plot_df(df, x, y, title="", xlabel='Date', ylabel='Value', dpi=100):
    # set plot figure size 16,5 and dpi as 100
    plt.figure(figsize=(16,5), dpi=100)
    # set title name, xlabel and ylabel
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)

    # plot using plot() method
    plt.plot(x,y)

    # set title name, xlabel and ylabel

    # show the plot
    plt.show()


# call the plot_df function and title as title='Monthly anti-diabetic drug sales in Australia from 1992 to 2008.')
plot_df(df=df, x=df.index, y=df['value'], title='Monthly anti-diabetic drug sales in Australia from 1992 to 2008.')
```

From the above plot you can see that the time series data that we have used has an increasing trend over the years. It also has seasonality as there are similar periodic changes in the data.<br><br> In the above plotting code, you saw the parameter dpi which is dots per inches. It represents the number of pixels per inch in the figure. The default value for dpi in matplotlib.pyplot.figure() function is 100.
